```python
# kilobytes = 1024 bytes
# megabytes = 1024*1024 bytes
def sizeInBytes(x):
    if(x.endswith("k")):
        number = x[0:-1]
        return float(number)*1024
    elif(x.endswith("M")):
        number = x[0:-1]
        return float(number)*1048576
        
```

```python
df["Size in bytes"] = df["Size"].apply(sizeInBytes)
```

```python
df
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
df
```

```python
X_part1 = df[['Reviews','Size in bytes','Numeric_install']]
```

First off I need to sort out the category column. This column contains string values from the list below. Seeing as this string value can only come from this list I can convert them into 1's and 0's to be used for my model as the category could be usfull for predicting the rating and i do not want to drop it.

the same can be said for the "Content Rating" column

```python
df['Category'].value_counts()
```

```python
X_part2 = df['Category']
```

```python
X_part2 = pd.get_dummies(X_part2)
```

```python
X_part2
```

```python
df['Content Rating'].value_counts()
```

```python
X_part3 = df['Content Rating']
```

```python
X_part3 = pd.get_dummies(X_part3)
```

```python
X_part3
```

```python
X = pd.concat([X_part1,X_part2,X_part3], axis=1 )
```

```python
y = df['Rating']
```

```python
X
```

```python
train_rmse =[]
test_rmse = []
```

```python
for d in range(1,6):
    
    model = PolynomialFeatures(degree =d)
    Extended_X = model.fit_transform(X)
    
    X_train, X_test, y_train, y_test = train_test_split(Extended_X, y, test_size=0.3, random_state=101)
    
    my_model = LinearRegression()
    my_model.fit(X_train, y_train)
    
    train_pred = my_model.predict(X_train)
    test_pred = my_model.predict(X_test)
    
    train_rmse.append(np.sqrt(mean_squared_error(y_train, train_pred)))
    test_rmse.append(np.sqrt(mean_squared_error(y_test, test_pred)))
```

I get ram allocation errors if i go past 6

```python
train_rmse
```

```python
test_rmse
```

```python
plt.figure(figsize=(6,4), dpi=150)
plt.plot(range(1,6), train_rmse[:5], label = 'Train RMSE')
plt.plot(range(1,6), test_rmse[:5], label = 'Test RMSE')
plt.xlabel('Degree of Polynomial')
plt.legend()
plt.ylim(0.34,0.4)
```

```python
model = PolynomialFeatures(degree =1) # using the above plot i can see that the test RMSE is lowest at 1
Extended_X = model.fit_transform(X)
    
X_train, X_test, y_train, y_test = train_test_split(Extended_X, y, test_size=0.3, random_state=101)
    
my_model = LinearRegression()
my_model.fit(X_train, y_train)
    
y_hat = my_model.predict(X_test)
    
MAE = mean_absolute_error(y_test, y_hat)
RMSE = np.sqrt(mean_squared_error(y_test, y_hat))
```
