```python
def extractYear(x):
    return x[:4]
```

# D

```python
df = df.drop(['title','speaker','recorded_date','published_date','event'], axis=1)
```

```python
df
```

```python
df = df.set_index('talk_id',verify_integrity=True)
```

I set the "talk_id" as the index as it is Unique for each talk

```python
df
```

```python
X_part1 = df[['duration', 'likes']]
```

```python
X_part2 = df['year']
X_part2 = pd.get_dummies(X_part2)
```

```python
X_part2
```

```python
X = pd.concat([X_part1,X_part2], axis=1 )
```

```python
X
```

```python
y = df['views']
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
```

```python
for d in range(1,2):
    
    model = PolynomialFeatures(degree =d)
    Extended_X = model.fit_transform(X)
    
    X_train, X_test, y_train, y_test = train_test_split(Extended_X, y, test_size=0.2, random_state=101) 
    
    # since the dataset will be devided into three parts, 20% for the test set is considered more than enough
       
    my_model = LinearRegression()
    scores = cross_val_score(my_model,X_train,y_train, cv=10, scoring ="neg_mean_squared_error"  )
    
```

```python
scores
```

```python
s = abs(scores)
s
```

I got both luck and unlucky with my partitions. 
Where i got a mean square error of only ~1: this was a lucky partition.
However where i get a mean squared error of ~5 or ~6: this was an unlucky partition

```python
s = np.sqrt(s)
```

```python
s
```

```python
s =s.mean()
s
```

```python
my_model.fit(X_train,y_train)
```

```python
test_pred = my_model.predict(X_test)
```

```python
test_rmse=(np.sqrt(mean_squared_error(y_test, test_pred)))
```

```python
test_rmse # here is the final unbiased error that can be reported 
```

I would not trust this model at all. The RMSE value should be <0.3 to be considered a good model. Here however the RMSE is well above this threshold at a stagering 64759.  This means that our views can be over or under by ~64000 views whic may not be a lot if the view are expecting to be in the millions but there are a few ted talks with only a few hundred views and this error can give us very wrong answers.
